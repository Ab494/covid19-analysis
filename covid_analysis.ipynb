  Options for handling Missing data:
1: Remove rows with missing values
2: Fill Missing values with 0 or a statistical value


Step 5: Handle the missing values by using Forward-Fill Vaccination Totals(Per Country)

```python
df.sort_values(by=['country', 'date'], inplace=True)

df['total_vaccinations'] = df.groupby('country')['total_vaccinations'].fillna(method='bfill')
df['people_fully_vaccinated'] = df.groupby('country')['people_fully_vaccinated'].fillna(method='bfill')
df['people_vaccinated'] = df.groupby('country')['people_vaccinated'].fillna(method='ffill')
df['daily_vaccinations_raw'] = df.groupby('country')['daily_vaccinations_raw'].fillna(0)
df['total_vaccinations_per_hundred'] = df.groupby('country')['total_vaccinations_per_hundred'].fillna(method='ffill')
df['people_vaccinated_per_hundred'] = df.groupby('country')['people_vaccinated_per_hundred'].fillna(method='ffill')
df['people_fully_vaccinated_per_hundred'] = df.groupby('country')['people_fully_vaccinated_per_hundred'].fillna(method='ffill')
df['daily_vaccinations_per_million'] = df.groupby('country')['daily_vaccinations_per_million'].fillna(method='ffill')
```

```python
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(0)
```

```python
df.isnull().sum()
```

Step 6: Top Countries by Total Vaccinations

```python
top_vaccinated = df.groupby('country')['total_vaccinations'].max().sort_values(ascending=False).head(10)
top_vaccinated.plot(kind='barh',  figsize=(10,6))
plt.title("Top 10 Countries by Total Vaccinations")
plt.xlabel("Total Vaccinations")
plt.gca().invert_yaxis()
plt.show()
```

- China has administered the highest number of covid-19 vaccine doses, standing out significantly from other countries
- The top  5 countries China, India, United States, Brazil and Indonesia are also among the most populous, explaining their totals
- Japan, Bangladesh, Pakistan, Vietnam, Mexico also appear in top 10, showing significant effort despite lower population sizes

Step 7: Fully Vaccinated People Per Hundred

```python
top_fully_vax = df.groupby('country')['people_fully_vaccinated_per_hundred'].max().sort_values(ascending=False).head(10)
top_fully_vax.plot(kind='bar', figsize=(10,6))
plt.title("Top 10 Countries by Full Vaccination Per Hunderd")
plt.ylabel("People Fully Vaccinated per 100")
plt.xticks(rotation=45)
plt.show()
```

- Gibralter leads significantly with 0ver 120 people fully vaccinated per 100,Gibralter shows the impact of booster campaigns and smaller population size
- High Perfoming small Countries like Pitcairn, Brunei and Cayman Islands demonstrate high per capita vaccination rates
- Most Countries in this chart have vaccinated more than their total population

Step 9: Daily Vaccination Distribution

```python
df['daily_vaccinations'].hist(bins=50, figsize=(10,6))
plt.title("Distribution of Daily Vaccinations")
plt.xlabel("Doses Administered")
plt.ylabel("Frequency")
plt.show()
```

- Most Days had a relatively low number of doses administered

- A few extreme values on the right suggest that some coutries had exceptionally high vaccination days these are rare but significant
